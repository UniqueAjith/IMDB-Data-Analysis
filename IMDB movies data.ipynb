**Groupby Operation**

```python
movie_data.groupby('Director')[['Rating']].mean().head()
```

```python
movie_data.groupby('Director')[['Rating']].mean().sort_values(['Rating'], ascending=False).head()
```

**To find Top Movies**

```python
def rating_group(rating):
    if rating >= 7.5:
        return 'Good'
    elif rating >= 6.0:
        return 'Average'
    else:
        return 'Bad'
```

```python
movie_data['Rating_category'] = movie_data['Rating'].apply(rating_group)
```

```python
movie_data[['Title','Director','Rating','Rating_category']].head(10)
```
